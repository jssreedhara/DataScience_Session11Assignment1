```python
# Assignment 11 by Sreedhara Jagatagar  Sreenivasa
#It happens all the time: someone gives you data containing malformed strings, Python,
#lists and missing data. How do you tidy it up so you can get on with the analysis?
#Take this monstrosity as the DataFrame to use in the following puzzles:
#df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
#'Budapest_PaRis', 'Brussels_londOn'],
#'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
#'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
#'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
#'12. Air France', '"Swiss Air"']})

#import required Libraray
import pandas as pd
import numpy as np

#Construct the Data Frame using above
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

#Print the data
df
```

```python
#Assignment 11 by Sreedhara Jagatagar  Sreenivasa

#1. Some values in the the FlightNumber column are missing. These numbers are meant
#to increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in
#these missing numbers and make the column an integer column (instead of a float column).

#Fill With 0
df['FlightNumber'].fillna(0, inplace=True)

#Write a Function to Increment previous values with 10
def custFill():
    previous=0
    for i in range(df.FlightNumber.count()):
        if df.FlightNumber.values[i]==0:
            df.FlightNumber.values[i]=previous
        else:
            previous=df.FlightNumber.values[i]+10

#Call the function to fill missing values
custFill() 

#Coververt Float values into Integer Values
df.FlightNumber=df.FlightNumber.astype(int)

#print the Dataframe
df
```

```python
#Assignment 11 by Sreedhara Jagatagar  Sreenivasa
#2. The From_To column would be better as two separate columns! Split each string on
#the underscore delimiter _ to give a new temporary DataFrame with the correct values.
#Assign the correct column names to this temporary DataFrame.

#Assign Dataframe to new temporay data frame.
df_temp=df

#Split the From and assign using _ Delimiter
df_temp['From']=df_temp.From_To.str.split('_').str[0]

#Split the To and assign using _ Delimiter
df_temp['To']=df_temp.From_To.str.split('_').str[1]

#Print the Temprary Dataframe
df_temp
```

```python
#Assignment 11 by Sreedhara Jagatagar  Sreenivasa


#3. Notice how the capitalisation of the city names is all mixed up in this temporary
#DataFrame. Standardise the strings so that only the first letter is uppercase (e.g."londON" should become "London".)

#Capitilize the From Field
df_temp['From']=df_temp.From.str.capitalize()

#Capitilize the To Field
df_temp['To']=df_temp.To.str.capitalize()

#Print the dataframe
df_temp
```

```python
#Assignment 11 by Sreedhara Jagatagar  Sreenivasa
#4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.

#Drop From_To Field from Temporary Data Frame.
df_temp.drop(['From_To'],axis=1, inplace=True)

#Assing or Attach the dataframe to original one.
df=df_temp

#print the dataframe
df
```
